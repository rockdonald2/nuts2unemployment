```python
import pandas as pd
import numpy as np
from pymongo import MongoClient
import requests
```

```python
# 'http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/
# nama_10_gdp?geo=EU28&precision=1&na_item=B1GQ&unit=CP_MEUR&time=2018&time=2019')
# <tablazat_nev>?geo=<terulet_filter>&<kulonbozo_filterek>&time=<idoszak>
```

```python
# egy algoritmus, amely megszerzi az adatokat az EUROSTAT API-ról, 2015-től 2019-ig

data = {}
year = 2015

for i in range(5):
    response = requests.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/lfst_r_lfu3rt?unit=PC&age=Y20-64&sex=T&time=' + str(year))
    json = response.json()
    
    geo_values = json['dimension']['geo']
    cell_values = json['value']
    
    for t in geo_values['category']['index']:
        label = t
        key = str(geo_values['category']['index'][t])
        nuts = geo_values['category']['label'][t]
        
        if key in cell_values:
            if not t in data:
                data[t] = {'region': nuts, str(year): cell_values[key]}
            else:
                data[t][str(year)] = cell_values[key]
    
    year = year + 1
```

```python
df = pd.DataFrame(data)
df = df.T

df.set_index(pd.Index(np.arange(0, len(data))))
df.reset_index(inplace=True)
df.rename(columns={'index': 'code'}, inplace=True)
```

```python
import re
```

```python
df = df.set_index(['code'])
```

```python
# elvetjük mindazon sorokat, amely kódjában nincsen szám, azok NUTS1 régiók általában, nekünk arra nincs szükségünk

indices = df.index.values

for i in indices:
    if re.search('\D$', i):
        df.drop(i, inplace=True)
        
df
```

```python
df[['2015', '2016', '2017', '2018', '2019']] = df[['2015', '2016', '2017', '2018', '2019']].where(df[['2015', '2016', '2017', '2018', '2019']].notnull(), np.nan)
df[['2015', '2016', '2017', '2018', '2019']] = df[['2015', '2016', '2017', '2018', '2019']].astype('float64')
```

```python
df.reset_index(inplace=True)
```
