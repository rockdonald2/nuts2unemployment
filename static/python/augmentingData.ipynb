```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sb
import requests
```

```python
df = pd.read_json('../data/nuts-data.json', orient='records')
df
```

```python
# hozzá szeretném előszöris rakni a régió teljes lakosságát, majd az aktív lakosságot, hogy képes legyek
# összehasonlításokat végezni
```

```python
# átnevezzük a munkanélküliségi rátákat tartalmazó oszlopokat e_<ev> sémára
year = 2015

for i in range(0, 5):
    df = df.rename(columns={'{}'.format(year): 'e_{}'.format(year)})
    
    year = year + 1
    
df
```

```python
# hozzáadjuk az új p_<ev> oszlopokat, amelyek tartalmazni fogják a teljes lakossági adatokat

year = 2015

for i in range(0, 5):
    df.insert(loc=(i + 1), column='p_{}'.format(year), value=np.nan)
    
    year = year + 1
    
df
```

```python
# hozzáadjuk az új a_<ev> oszlopokat, amely az aktív lakosság adatait fogja tartalmazni

year = 2015

for i in range(0, 5):
    df.insert(loc=(i + 1), column='a_{}'.format(year), value=np.nan)
    
    year = year + 1
    
df
```

```python
# átrendezzük az oszlopok sorrendjét, code-region-adatok-ra

cols = df.columns.tolist()
cols = cols[1:11] + cols[12:]
cols.insert(0, 'code')
cols.insert(1, 'region')
cols

df = df[cols]
```

```python
response = requests.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/lfst_r_lfsd2pwn?age=Y20-64&sex=T&time=2019&geo=AT11&wstatus=ACT&citizen=TOTAL')
```

```python
response.json()
```

```python
df_code_index = df.set_index('code')
```

```python
codes = df_code_index.index.values # lekérjük a NUTS2 regiókódokat
```

```python
# megszerezzük a teljes lakosságra szóló adatokat, és behelyezzük a dataframe-ünkbe

year = 2015

for i in range(0, 5):
    for c in codes:
        resp = requests.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/lfst_r_lfsd2pwn?age=Y20-64&sex=T&time={}&geo={}&wstatus=POP&citizen=TOTAL'.format(year, c))
        df_code_index.loc[c, 'p_{}'.format(year)] = resp.json()['value']['0']
    
    year = year + 1
    
df_code_index
```

```python
# megszerezzük az aktív lakosságra szóló adatokat, és behelyezzük a dataframe-ünkbe

year = 2015

for i in range(0, 5):
    for c in codes:
        resp = requests.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/lfst_r_lfsd2pwn?age=Y20-64&sex=T&time={}&geo={}&wstatus=ACT&citizen=TOTAL'.format(year, c))
        df_code_index.loc[c, 'a_{}'.format(year)] = resp.json()['value']['0']
    
    year = year + 1
    
df_code_index
```

```python
# általánosságban elmondható, hogy az e = munkanélküliek száma / aktív lakosság számával
# a high-low tooltip rész alatt, azt értem, hogy lesz egy olyan event, amikoris rászviszem az egeret
# az adott ország területére, és a tooltip-ban kiírja, hogy 
# átlagviszonylatban magas vagy alacsony-e az aktív lakosság/teljes lakosság
```

```python
df = df_code_index.reset_index()
df
```

```python
df.to_json('../data/nuts-data.json', orient='records')
```
