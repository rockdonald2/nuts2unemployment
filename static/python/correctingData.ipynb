```python
import pandas as pd
import numpy as np
import requests
```

```python
df = pd.read_json('../data/nuts-data.json', orient='records')
```

```python
new_df = pd.DataFrame(columns=df.columns.tolist())

new_df
```

```python
df_code_index = df.set_index('code')

for c in codes:
    try:
        df_code_index.loc[c]
    except KeyError:
        year = 2015
        
        e = {}
        p = {}
        a = {}
        
        FLAG = False
        
        for _ in range(5):
            resp_e = requests.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/lfst_r_lfu3rt?unit=PC&age=Y20-64&sex=T&time={}&geo={}'.format(year, c))
            resp_p = requests.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/lfst_r_lfsd2pwn?age=Y20-64&sex=T&time={}&geo={}&wstatus=POP&citizen=TOTAL'.format(year, c))
            resp_a = requests.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/lfst_r_lfsd2pwn?age=Y20-64&sex=T&time={}&geo={}&wstatus=ACT&citizen=TOTAL'.format(year, c))
            
            if not (resp_e.status_code == 200):
                FLAG = True
                break
            elif not (resp_p.status_code == 200):
                FLAG = True
                break
            elif not (resp_a.status_code == 200):
                FLAG = True
                break
            
            region = resp_e.json()['dimension']['geo']['category']['label'][c]

            if len(resp_e.json()['value']) > 0:
                e[str(year)] = resp_e.json()['value']['0']
            else:
                e[str(year)] = np.nan
                
            if len(resp_p.json()['value']) > 0:
                p[str(year)] = resp_p.json()['value']['0']
            else:
                p[str(year)] = np.nan
            
            if len(resp_a.json()['value']) > 0:
                a[str(year)] = resp_a.json()['value']['0']
            else:
                a[str(year)] = np.nan
            
            year = year + 1
            
        if FLAG:
            continue
        
        new_df = new_df.append({'code': c, 'region': region, 'a_2015': a['2015'], 'a_2016': a['2016'], 'a_2017': a['2017'], \
                   'a_2018': a['2018'], 'a_2019': a['2019'], \
                   'p_2015': p['2015'], 'p_2016': p['2016'], 'p_2017': p['2017'], 'p_2018': p['2018'], 'p_2019': p['2019'], \
                  'e_2015': e['2015'], 'e_2016': e['2016'], 'e_2017': e['2017'], 'e_2018': e['2018'], 'e_2019': e['2019']
                  }, ignore_index=True)
```
